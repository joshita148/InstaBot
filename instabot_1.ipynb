## 1.	Login to your Instagram Handle

```python
def Login(username, password):
    
    #finding username box and sending username to it
    user_name= wait.until(EC.presence_of_element_located((By.NAME, "username")))   
    user_name.send_keys(username)
    
    #finding password box and sending password to it
    pass_word= driver.find_element_by_name('password')
    pass_word.send_keys(password)
    
    #submitting/login in
    pass_word.submit()
    
    #closing 2 pop-ups by selecting not now option
    not_now= wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class= "cmbtv"]/button')))
    not_now.click()
    turn_on_notifications= wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class= "mt3GC"]/button[last()]')))
    turn_on_notifications.click()
    
```

## 2.	Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food”
#####  1.	Note : Make sure to avoid printing hashtags


```python
def search(s):
    names=[]
    
    #finding search box and sending string to be searched
    search= driver.find_element_by_xpath('//div[@class=" QY4Ed"]/input')
    search.send_keys(s)
    
    #wait until the dopbox appears
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'fuqBx ')))
    
    # finding usernames of all that appeared in the dropbox
    drpdwn= driver.find_elements_by_class_name('_7UhW9.xLCgt.qyrsm.KV-D4.uL8Hv')
    
    #appending all the username into a list excluding hashtags
    for profile in drpdwn:
        name= profile.text
        if name[0] != '#':
            names.append(name)
        
    #clearing the search box
    search.clear()
    
    return names
```

```python
name_list = search('food')

for i in name_list :
    print(i)
```

```python
def follow():
    
    #finding the follow button
    follow= driver.find_element_by_xpath('//button[starts-with(@class,"_5f5mN")]')
    
    if follow.text=='Follow':
        follow.click()
        print('Followed')
    else:
        print(' Already Following')
```

```python
def Unfollow():
    
    #finding the follow button
    button= driver.find_element_by_xpath('//button[starts-with(@class,"_5f5mN")]')
    
    if button.text!='Follow':
        button.click()
        
        #finding and clicking unfollow on the pop-up
        unfollow=  driver.find_element_by_class_name('aOOlW.-Cab_')
        unfollow.click()
        print('UnFollwed')
    else:
        print('Already Unfollowed')
```

## 5.	Like/Unlike posts
##### 1.	Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
##### 2.	Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.


```python
def like_unlike():

    #finding the search box and passing the string
    search= driver.find_element_by_xpath('//div[@class="_0aCwM "]/input')
    search.send_keys('dilsefoodie')
    
    #finding and clicking the hashtag
    dilsefoodie= wait.until(EC.presence_of_element_located((By.XPATH, '//a[@href= "/explore/tags/dilsefoodie/"]')))
    dilsefoodie.click()
    
    #opening the 1st post of the hashtag
    post1= driver.find_element_by_class_name('_9AhH0')
    post1.click()
    
    #liking top 30 posts
    for i in range(30):
        wait= WebDriverWait(driver, 10)
        #finding and clicking the like button
        like= wait.until(EC.presence_of_element_located((By.XPATH, '//div[contains(@class, "rrUvL")]')))
        like.click()
        time.sleep(1)

        #finding and clicking the next button
        next_= driver.find_element_by_xpath('//div[starts-with(@class, " l8mY4")]//div')
        next_.click()
        time.sleep(1)

    print('liked top 30 posts of "dilsefoodie"')
    
    #unliking the top 30 posts
    for i in range(30):
        #finding and clicking the previous button
        previous= driver.find_element_by_xpath('//div[starts-with(@class, "_6Eych")]//div')
        previous.click()
        time.sleep(1)

        #finding and clicking the like button to unike it
        unlike= wait.until(EC.presence_of_element_located((By.XPATH, '//section[contains(@class, "Slqrh")]//div[@class= "QBdPU "]')))
        unlike.click()
        time.sleep(1)
     

    print('Unliked ALL top 30 posts of "dilsefoodie"')
    
```

## 6.   Extract list of followers

#### 1.Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’

```python
def Extract_Followers():
    
    #finding and clicking the follwers option
    foodtalkindia=driver.find_element_by_partial_link_text('followers')
    foodtalkindia.click()
    
    #finding pop-up so that we can scroll in it
    popup = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[@class='isgrP']")))
  
    #scrolling
    for i in range(50):
        driver.execute_script('arguments[0].scrollTop = arguments[0].scrollTop + arguments[0].offsetHeight;', popup)
        time.sleep(1)

        
    names = []
    #extracting usernames of followers and appending it in the list
    followers = driver.find_elements_by_class_name('d7ByH')
    for i in followers[:500] :
            names.append(i.text.split('\n')[0])
            
    return names
```

#### 2.Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.


```python
def Following():

    #finding and clicking the following option
    following_btn = wait.until(EC.presence_of_element_located((By.PARTIAL_LINK_TEXT,'following')))
    following_btn.click()
        
    #finding the following pop-up
    frame = driver.find_element_by_class_name('isgrP')
    #scrolling 
    for i in range(25):
        time.sleep(1)
        driver.execute_script("arguments[0].scrollTop=arguments[0].scrollHeight",frame)
        
    names = []
    #extracting usernames and appending it in the list
    following = driver.find_elements_by_class_name('d7ByH')
    for i in following :
        names.append(i.text.split('\n')[0])
            
    return names
```

```python
def Followed_by():
    names=[]
    
    #finding the usernames mentioned in the line 'followed by...'
    following = wait.until(EC.presence_of_all_elements_located((By.XPATH,'//span[@class= "_32eiM"]')))
    
    for i in following:
            names.append(i.text)
            
    return names
```

```python
#for search and open 'foodtalkindia' instagram handle
search_open('foodtalkindia')

# Extracting followers using Extract_Followers() function
followers_of_foodind = Followed_by()

#casting into set
followers_of_foodind = set(followers_of_foodind)
```

```python
#finding handles that I follow
username= 'SAMPLE USERNAME'
search_open(username)
followed_by_me = Following()
followed_by_me = set(followed_by_me)
```

```python
#taking intersection so s1 contains only that users that are followed by me and who follow foodtalkindia 
s1=(followers_of_foodind).intersection(followed_by_me)
if len(s1) == 0:
    print('No such users found')
else:
    #now extracting my followers
    my_follower = Extract_Followers()
    my_follower = set(my_follower)
    #taking intersection with s1, so s2 contains only users that I am following them but they don’t follow me
    s2=s1.intersection(my_follower)
    if len(s2) == 0:
        print('No such users found')
    else:
        
        for user in s2:
            print(user)
```

## 7. Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly

```python
def Check_Story():
    try:
        #finding story or profile pic
        story = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"CfWVH")))
    
        #check the Profile photo size to find out story is available or not
        height = driver.find_element_by_class_name('CfWVH').get_attribute('height')
        if int(height) == 131:
            print("Already seen the story")
        else:
            print("Viewing the story")
            driver.execute_script("arguments[0].click();",story)
    except:
        print("No story is available to view")
```
